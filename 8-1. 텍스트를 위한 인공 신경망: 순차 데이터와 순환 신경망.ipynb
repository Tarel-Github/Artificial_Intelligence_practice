수 없이 많은 텍스트를 빠르게 분류할 수 있는 알고리즘을 작성해 보자  
# 순차 데이터
순차 데이터는 텍스트나 시계열 데이터와 같이 순서에 의미가 있는 데이터를 말한다.  
예를 들어 I am a boy는 말이 되지만 boy am a I는 말이 되지 않는다.  
지금까지의 데이터는 순서와는 상관이 없었지만  
텍스트 데이터에는 단어의 순서가 중요하게 된다.  
때문에 순차데이터를 다룰 때는 이전 데이터를 기억하는 기능이 필요하다.  
예를 들어, "별로지만 추천한다." 는 문자열에서, "별로"라는 데이터를 기억해야지만  
이 말이 무조건 긍정적인건 아니다 라는 것을 알 수 있다.  

---

입력 데이터의 흐름이 앞으로만 전달되는 신경망을 피드포워드 신경망 이라고 한다. 
다음 샘플을 위해서 이전 데이터가 신경망 층에 순환될 필요가 있는데  
이를 순환 신경망이라고 한다.  


# 순환 신경망
순환 신경망은 일반적인 완전 연결 신경망과 거의 비슷하다.  
완전연결 신경망에서 뉴런의 출력이 다시 자기 자신으로 전달된다.  
즉 어떤 샘플 데이터를 처리할 때, 바로 이전에 사용했던 데이터를 재사용하는 것이다.
