- 포워드 패스: 학습 예제를 네트워크에 제시하고 출력을 정답과 비교
- 백워드 패스: 가중치에 대한 편미분을 계산한다.  
  - 이러한 미분은 출력이 정답에 가까워지도록 가중치를 조정하는데 쓴다.
