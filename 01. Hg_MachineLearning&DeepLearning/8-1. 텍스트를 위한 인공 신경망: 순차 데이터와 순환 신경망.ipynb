# 순차 데이터
순차 데이터는 텍스트나 시계열 데이터와 같이 순서에 의미가 있는 데이터를 말한다.  
예를 들어 I am a boy는 말이 되지만 boy am a I는 말이 되지 않는다.  
지금까지의 데이터는 순서와는 상관이 없었지만  
텍스트 데이터에는 단어의 순서가 중요하게 된다.  
때문에 순차데이터를 다룰 때는 이전 데이터를 기억하는 기능이 필요하다.  
예를 들어, "별로지만 추천한다." 는 문자열에서, "별로"라는 데이터를 기억해야지만  
이 말이 무조건 긍정적인건 아니다 라는 것을 알 수 있다.  

---

입력 데이터의 흐름이 앞으로만 전달되는 신경망을 피드포워드 신경망 이라고 한다. 
다음 샘플을 위해서 이전 데이터가 신경망 층에 순환될 필요가 있는데  
이를 순환 신경망이라고 한다.  


# 순환 신경망
순환 신경망은 일반적인 완전 연결 신경망과 거의 비슷하다.  
완전연결 신경망에서 뉴런의 출력이 다시 자기 자신으로 전달된다.  
즉 어떤 샘플 데이터를 처리할 때, 바로 이전에 사용했던 데이터를 재사용하는 것이다.  
  
1번 데이터가 지나가면 1차 처리 데이터가 만들어진다.  
2번 데이터가 지나가면서 1차 처리 데이터도 같이 지나가면,    
2차 처리 데이터가 만들어진다.  
이런식으로 반복하면? 최초 데이터부터 끝 데이터까지 포함하게 된다.  

이런식으로 샘플을 처리하는 한 단계를 <b>타임스텝</b>이라고 한다.  

3차 처리데이터에는 1차 처리 데이터보다 1번 데이터의 양이 적다.  
때문에 순환을 거듭할 수록 데이터 최초 데이터가 희미해지는데 이에 대해선 추후에 후술한다.  
  
---
순환 신경망에는 층을 <b>셀</b>이라고 부른다.  
셀 하나에는 뉴런이 여러개가 있지만 완전연결신경망과는 다르게  
모든 뉴런을 표시하지 않고 하나의 셀로 층을 표현한다.  

---
일반적으로 은닉층의 활성화 함수는 하이퍼볼릭 탄젠트를 주로 쓰는데  
시그모이드 함수랑 비슷하게 생겼다.  
단, 시그모이드와는 달리 -1 ~ 1 까지의 범위를 가진다.  


# 셀의 가중치와 입출력
입력층 특성이 4개, 순환층 뉴런이 3개라고 가정해보자  
입력층과 순환층이 모두 연결되기 때문에 가중치는 4 * 3 = 12 가 된다.  
여기까진 완전연결 신경망과 같다.  
  
이제 순환층에서 타임스텝을 한다.  
순환층의 1번 뉴런이 첫번째 타임스텝에서, 1 ~ 3번 뉴런 모두에 자신의 상태가 전달된다.  
나머지 뉴런들도 마찬가지다.  
고로 이 순환 층의 은닉 상태를 위한 가중치 wb는 3 * 3 = 9 개다.  
  
자, 이제 파라미터는 몇 개인가?  
가중치에 절편을 더하면 된다.  
여기엔 각 뉴런마다 하나의 절편이 있다.  
고로 이 순환층은 모두 12 + 9 + 3개의 파라미터가 있다.  
  
이제 순환층의 입출력에 대해 살펴보자  
순환층은 일반적으로 샘플마다 2개의 차원을 가진다.
