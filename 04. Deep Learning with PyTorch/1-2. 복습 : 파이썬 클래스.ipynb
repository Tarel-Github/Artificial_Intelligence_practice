## 함수

함수는 일련의 작업을 수행하는 코드 블록이다.  
result 변수를 정의하고 add 라는 함수를 만들어 계산을 수행해보자  

</br>
함수는 add 와 add2를 만들 것이다.  

```python
result1 = 0
result2 = 0
```

```python
def add1(num):
    global result1
    result1 += num
    return result1
```

```python
def add2(num):
    global result2
    result2 += num
    return result2
```

```python
print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))
```

서로의 값에 영향을 주지 않고 서로 다른 연산을 하고 있다.  
이번엔 클래스로 구현해보자

## 클래스
클래스는 함수를 하나로 묶어놓은 사용자 정의 데이터 타입이다.  
클래스는 캡슐화와 정보은닉과 추상화를 제공한다.  

```python
class Calculator:
    def __init__(self): # 객체 생성 시 호출될 때 실행되는 초기화 함수. 이를 생성자라고 한다.
        self.result = 0

    def add(self, num): # 객체 생성 후 사용할 수 있는 함수.
        self.result += num
        return self.result
```

```python
cal1 = Calculator()
cal2 = Calculator()
```

클래스를 선언하고 객체 2개를 만들었다.

```python
print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```

역시 독립적으로 연산되는 것을 볼 수 있다.
