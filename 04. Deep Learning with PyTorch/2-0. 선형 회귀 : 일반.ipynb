# 선형 회귀
선형 회귀에 대해 이해하고 선형회귀 모델을 만들어 보자  
아래 4가지에 대해 알아볼 생각이다.  

</br>

#### ● 데이터에 대한 이해
*   학습할 데이터에 대해 알아보자

#### ● 가설 수립
*   가설을 수립하는 방법에 대해 알아보자

#### ● 손실 계산
*   학습 데이터를 이용해서 연속적으로 모델을 개선시키는데 손실을 계산해보자

#### ● 경사 하강법
*   학습 핵심 알고리즘인 경사하강법에 대해 알아보자
