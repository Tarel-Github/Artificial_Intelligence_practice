# 가설 수립

머신러닝에서 식을 세울 때, 이 식을 가설이라고 한다.  
선형 회귀 가설은 이미 널리 알려져 있으므로 고민할거 없다.  
선형회귀 방정식은 아래와 같은 형식을 가진다.  

```py
y= Wx+b
# 아래 처럼 표기하기도 한다.
H(x)= Wx+b
```
그냥 중학생때 배운 선형 방정식이다.

점 4개는 훈련 데이터다.

뭐가 가장 잘 나타낸 직선인지 선택하려면 근거가 있어야 한다.  
그래서 오차라는 개념을 도입한다.  

image.png

실제 데이터 값과 빨간 선이 가지는 차이를 화살표로 표현했다.  
빨간 선은 y = 13x + 1이다.  

</br>

각 오차는 음수도 있고 양수도 있다.  
각 오차를 모두 제곱해서 양수로 만든 뒤에 전부 더한다.  
더한 결과는 다음과 같다.  
```
(-2)^2 + 10^2 + (-9)^2 + (-5)^2 = 210
```
이 때, 데이터가 4개이므로 4로 나누면 평균을 구할 수 있다.  
평균은 다음과 같다.  
```
210 / 4 = 52.5
```
우리의 목표는 이 값을 최소로 하는 선을 구하는 것이다.  
이것을 구하는 것이 옵티마이저 알고리즘(최적화 알고리즘) 이며  
이 알고리즘을 통해 적절한 W와 b를 찾는 과정을  
머신러닝에서 학습 이라고 부른다.  

</br>

여기서는 가장 기본적인 옵티마이저 알고리즘인  
경사하강법에 대해서 알아보자
