<h2>퍼셉트론</h2>

퍼셉트론은 직선방정식을 사용하여 선형회귀와 유사한 구조를 가진다.  
그러나 마지막에서 계단함수를 사용한다.  
그리고 계단 함수를 통과한 값을 다시 가중치와 절편을 업데이트  
즉, 학습시키는데 사용한다.  
다음 수식을 선형 함수라고 해보자  
  

```
w1·x1 + w2·x2 + b = z
```
여기서 계단 함수는 z가 0 이상이면 1, 작으면 -1로 분류한다.  
즉 다음 연산을 진행한다.  

```C
z >= 1 ? y = 1 : y = -1
```

이때 1을 양성클래스, -1을 음성클래스라고 부르며 함수로 나타내면 계단 모양이 나온다.  
때문에 계단함수라고 부른다.  
  
요약하자면 선형 함수를 통과한 값 z를 계단함수로 보내 0보다 큰지 작은지를  
검사하여 -1 또는 1로 분류하여 가중치와 절편을 업데이트 하는 알고리즘이다.  
  

앞서 보여준 선형 함수의 밑첨자는는 n번째 특성의 가중치와 입력을 의미한다.  
즉, 특성이 n개인 선형함수는 다음과 같이 표기할 수 있다.  

```
z = w1·x1 + w2·x2 ...... wn·xn + b
```

### 로지스틱 회귀

로지스틱 회귀는 아달린에서 조금 더 발전한 형태를 가진다.  
로지스틱 회귀는 선형 함수를 통과하여 얻은 z를 활성화 함수에 보낸다.  
활성화 함수를 통과한 값은 a라고 하면 이 a를 다시 임계함수에 넣는다.  
임계함수란 계단함수와 비슷하지만 z대신 a를 사용한다는 차이만 있을 뿐이다.  

정리하면 다음과 같다.  
```
w1x1, w2x2, b를 선형함수에 대입

선형함수 -> z값 -> 활성화 함수 -> a값 -> 임계함수 -> y

여기서 a를 역방향 계산에 사용
```
보다 더 자세한 내용은 실습을 진행하면서 살펴보자  
  
</br>

### 활성화 함수

활성화 함수는 보통 비선형 함수를 사용한다.  
왜 비선형 함수를 사용할까?  
</br>

예를 들어, 선형 함수 a = w1x1 + w2x2 ... wnxn가 있고, 활성화 함수로  
y=ka와 같은 선형함수를 사용한다고 가정해보면 결국  
y = k(w1x1 + ...... + wnxn)이 될 것이다.  
이렇게 되면 임계 함수 앞에 뉴런을 여러 개 쌓아도 결국 선형함수이니  
별 의미가 없다. 
  
때문에 활성화 함수는 의무적으로 비선형 함수를 사용한다.  
  
로지스틱 회귀에는 <b>'시그모이드 함수'</b>를 활성화 함수로써 사용한다.  

# 시그모이드 함수

시그모이드 함수는 z를 0~1 사이의 확률 값 a로 변환시켜준다.  
일반적으로, a가 0.5 이상이면 양성클래스, 아니면 음성클래스로 구분한다.  

## 오즈비
시그모이드 함수는 오즈비라는 통계를 기반으로 만들어지며  
다음과 같이 정의할 수 있다.  


```
OR = p / (p-1)

p = 성공확률
```

## 로짓함수
오즈비에 로그 함수를 취해서 만든 함수다.  
다음과 같이 정의할 수 있다.

```
logit(p) = log(OR)

OR은 선술한 오즈비
```

로짓 함수는 p가 0.5일 때 0이되고 0과 1일때 각각 무한대로 음수와 양수가 된다.  

이 식은 다음과 같이 쓸 수 있다.

```
log(p / (1 - p)) = z
```

## 로지스틱 함수

위 식을 다시 z에 대하여 정리하면 다음과 같은 식이 된다.

```
p = 1 / (1 + e ^(-1))
```
이게 로지스틱 함수다.  

로지스틱 함수를 그래프로 그리면 로짓 함수의 가로세로가 뒤집힌 형태다.  
로지스틱 함수를 시그모이드 함수라고도 부른다.

로지스틱 회귀는 이진분류가 목표다.

즉 이진 분류가 된 것이다.  
  
이제 가중치와 절편을 적절히 업데이트 하는 방법,  
로지스틱 회귀에 적용할 손실함수에 대해서 알아본다.

# 로지스틱 손실 함수
로지스틱 손실 함수는 크로스 엔트로피 손실 함수를 이진분류 버전으로 만든 것이다.  

이것을 식으로 나타내면 아래와 같다.  
a는 활성화 함수가 출력한 값이고 y는 타깃이다.  
```
L = -(ylog(a) + (1 - y)log(1 - a))
```

이진분류는 0과 1로 분류하는 함수로 답은 오직 둘 중 하나만 나온다.  
따라서 위 식은 y가 0이냐 1이냐로 정리된다.  
  
  
## 로지스틱 손실 함수 미분
  
로지스틱 손실함수를 미분한 식은 가중치와 절편 업데이트에 사용된다.

```
가중치의 경우 = - (y - a)x
절편의 경우   = - (y - a)1
```

로지스틱 손실함수의 미분을 통해 로지스틱 손실함수의 값을 최소로 하는  
가중치와 절편을 찾아야 한다.

## 로지스틱 손실함수와 연쇄법칙

예를 들어 다음과 같은 함수가 있다.
```
y = f(u)
u = g(x)

정리하면 y = f(g(x))
```
여기서 y를 x에 대해 미분하는 방법은 y를 u에 대해 미분한 값과 u를 x에 대해 미분한 값을 곱하는 것이다.  
즉 아래와 같다.  

```
y(x에 미분) = y(u에 미분) * u를(x에 미분)
```
  
만약 로지스틱 손실함수 L을 가중치w나 절편 b에 대해서 바로 미분하면 너무 복잡해진다.  

로지스틱 손실 함수L을  활성화 함수의 출력값 a에 대해 미분하고  
출력값 a는 선형함수의 출력값 z에 대해 미분하고  
선형함수의 출력값 z는 가중치 w 또는 절편 b에 대하여 미분한 다음 서로 곱한다.  
  
이러면 결국 이루가 원하는 로지스틱 손실함수를 가중치에 대해 미분한 값을 얻을 수 있다.  
또 이 과정은 역방향으로 진행된다는 것도 알 수 있다.  

### a를 z에 대해서 미분 
a는 시그모이드 함수고 z는 선형함수다.  
a를 z에 대해서 미분하면 결과적으로 다음과 같이 정리된다.  
```
a(z에 대해 미분) = a(1 - a)
```

### z를 w에 대해서 미분하기

z는 선형함수기 때문에 미분하면 다른 모든건 다 사라지고 x만 남는다.  

### 로지스틱 손실함수를 w에 대하여 미분하기

이제 필요한 계산은 모두 다 했으니 정리해보자
```
L(w에 대해 미분) = -(y - a)x
```
즉, 로지스틱 손실 함수를 w에 대해 미분한 결과는 제곰오차를 미분한 결과와 일치한다.  















## 가중치 업데이트
가중치 업데이트는 로지스틱 손실함수를 가중치에 대해 미분한 식을 가중치에서 빼면 된다.  
```
w(새 가중치) = w(기존 가중치) - L(기존 가중치에 대해 미분)
```

## 절편 업데이트
절편 업데이트는 로지스틱 손실함수를 <b>절편</b>에 대해 미분한 식을 빼면 된다.  
```
b = b - L(기존 절편에 대해 미분)
```

# 로지스틱 회귀 구현
준비된 데이터를 바탕으로 로지스틱 회귀를 구현해 보자

```python
class LogisticNeuron:
  def __init__(self):                 # 이전과는 달리 가중치와 절편을 미리 초기화 하지 않는다.
    self.w = None
    self.b = None
  
  def forpass(self, x):
    z = np.sum(x * self.w) + self.b   # 직선방정식 계산, 넘파이 함수를 사용한다.
    return z

  def backprop(self, x, err):
    w_grad = x * err                  # 가중치에 대한 그레이디언트를 계산     그레이디언트 = 손실함수의 변화율
    b_grad = 1 * err                  # 절편에 대한 그레이디언트를 계산
    return w_grad, b_grad

```

넘어가기 전에 넘파이 배열에 잠시 알아보고 가자

````python
# 넘파이 배열 특성곱에 대해 잠시 알아보고 가자
a = np.array([1, 2, 3])
b = np.array([3, 4, 5])
print(a+b)    # 각 요소를 더한 값 반환
print(a*b)    # 각 요소를 곱한 값 반환
np.sum(a*b)   # 각 요소를 곱한 뒤, 더한 값 반환
````

# 훈련 메서드 구현
훈련 메서드는 보통 fit()이다.  
이걸 구현해보자

```python
def fit(self, x, y, epochs = 100):
  self.w = np.ones(x.shape[1])              # 가중치 초기화
  self.b = 0                                # 절편을 초기화
  for i in range(epochs):                         # epochs 만큼 반복한다.
    for x_i, y_i in zip(x, y):                    # 모든 샘플에 대해 반복한다.
      z = self.forpass(x_i)                       # 정방향 계산
      a = self.activation(z)                      # 활성화 함수 적용
      err = -(y_i -a)                             # 오차 계산
      w_grad, b_grad = self.backprop(x_i, err)    # 역방향 계산
      self.w -= w_grad                                # 가중치 업데이트
      self.b -= b_grad                                # 절편 업데이트
```

# 활성화 메서드 구현
시그모이드 함수를 사용한다.  
시그모이드 함수는 넘파이의 np.exp()함수를 사용하면 간단히 만들 수 있다.  

```python
# 시그모이드 함수
def activation(self, z):
  a = 1 / (1 + np.exp(-z))
  return a
```

# 예측 메서드 구현
predict() 메서드를 구현하자  
입력값 x가 2차원 배열로 전달된다고 가정하고 구현한다.  
  
예측값은 입력값을  
선형함수 -> 활성화함수 -> 임계함수 순서로 통과시키면 된다.  
이건 전부 앞서 구현했으니, 간단하게 구현할 수 있다.  

```python
def predict(self, x):
  z = [self.forpass(x_i) for x_i in x]
  a = self.activation(np.array(z))
  return a > 0.5
```

# 코드 합치기
지금까지 작성한 코드를 하나로 합치자

````python
class LogisticNeuron:
  def __init__(self):                 # 이전과는 달리 가중치와 절편을 미리 초기화 하지 않는다.
    self.w = None
    self.b = None
  
  def forpass(self, x):
    z = np.sum(x * self.w) + self.b   # 직선방정식 계산, 넘파이 함수를 사용한다.
    return z

  def backprop(self, x, err):
    w_grad = x * err                  # 가중치에 대한 그레이디언트를 계산     그레이디언트 = 손실함수의 변화율
    b_grad = 1 * err                  # 절편에 대한 그레이디언트를 계산
    return w_grad, b_grad

  def fit(self, x, y, epochs = 100):
    self.w = np.ones(x.shape[1])              # 가중치 초기화
    self.b = 0                                # 절편을 초기화
    for i in range(epochs):                         # epochs 만큼 반복한다.
      for x_i, y_i in zip(x, y):                    # 모든 샘플에 대해 반복한다.
        z = self.forpass(x_i)                       # 정방향 계산
        a = self.activation(z)                      # 활성화 함수 적용
        err = -(y_i -a)                             # 오차 계산
        w_grad, b_grad = self.backprop(x_i, err)    # 역방향 계산
        self.w -= w_grad                                # 가중치 업데이트
        self.b -= b_grad                                # 절편 업데이트

  # 시그모이드 함수
  def activation(self, z):
    a = 1 / (1 + np.exp(-z))
    return a

  def predict(self, x):
    z = [self.forpass(x_i) for x_i in x]
    a = self.activation(np.array(z))
    return a > 0.5
````
