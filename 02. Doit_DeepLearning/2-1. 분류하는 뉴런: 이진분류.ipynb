<h2>퍼셉트론</h2>

퍼셉트론은 직선방정식을 사용하여 선형회귀와 유사한 구조를 가진다.  
그러나 마지막에서 계단함수를 사용한다.  
그리고 계단 함수를 통과한 값을 다시 가중치와 절편을 업데이트  
즉, 학습시키는데 사용한다.  
다음 수식을 선형 함수라고 해보자  
  

```
w1·x1 + w2·x2 + b = z
```
여기서 계단 함수는 z가 0 이상이면 1, 작으면 -1로 분류한다.  
즉 다음 연산을 진행한다.  

```C
z >= 1 ? y = 1 : y = -1
```

이때 1을 양성클래스, -1을 음성클래스라고 부르며 함수로 나타내면 계단 모양이 나온다.  
때문에 계단함수라고 부른다.  
  
요약하자면 선형 함수를 통과한 값 z를 계단함수로 보내 0보다 큰지 작은지를  
검사하여 -1 또는 1로 분류하여 가중치와 절편을 업데이트 하는 알고리즘이다.  
  

앞서 보여준 선형 함수의 밑첨자는는 n번째 특성의 가중치와 입력을 의미한다.  
즉, 특성이 n개인 선형함수는 다음과 같이 표기할 수 있다.  

```
z = w1·x1 + w2·x2 ...... wn·xn + b
```

## 아달린

퍼셉트론을 개선한 적응형 선형 뉴런을 아달린이라고 부른다.  
아달린은 선형함수의 결과를 학습에 사용한다.  
계단 함수의 결과는 예측에만 활용한다.  
  
퍼셉트론에서는 z를 계단함수에 넣어서 y값을 구했다.  

하지만 여기서는 z를 계단함수에 넣기 전에 역방향 계산을 수행한다.  
단, z값은 계단함수에도 그대로 들어간다.

### 로지스틱 회귀

로지스틱 회귀는 아달린에서 조금 더 발전한 형태를 가진다.  
로지스틱 회귀는 선형 함수를 통과하여 얻은 z를 활성화 함수에 보낸다.  
활성화 함수를 통과한 값은 a라고 하면 이 a를 다시 임계함수에 넣는다.  
임계함수란 계단함수와 비슷하지만 z대신 a를 사용한다는 차이만 있을 뿐이다.  

정리하면 다음과 같다.  
```
w1x1, w2x2, b를 선형함수에 대입

선형함수 -> z값 -> 활성화 함수 -> a값 -> 임계함수 -> y

여기서 a를 역방향 계산에 사용
```
보다 더 자세한 내용은 실습을 진행하면서 살펴보자  
  
</br>

### 활성화 함수

활성화 함수는 보통 비선형 함수를 사용한다.  
왜 비선형 함수를 사용할까?  
</br>

예를 들어, 선형 함수 a = w1x1 + w2x2 ... wnxn가 있고, 활성화 함수로  
y=ka와 같은 선형함수를 사용한다고 가정해보면 결국  
y = k(w1x1 + ...... + wnxn)이 될 것이다.  
이렇게 되면 임계 함수 앞에 뉴런을 여러 개 쌓아도 결국 선형함수이니  
별 의미가 없다. 
  
때문에 활성화 함수는 의무적으로 비선형 함수를 사용한다.  
  
로지스틱 회귀에는 <b>'시그모이드 함수'</b>를 활성화 함수로써 사용한다.  

# 시그모이드 함수

시그모이드 함수는 z를 0~1 사이의 확률 값 a로 변환시켜준다.  
일반적으로, a가 0.5 이상이면 양성클래스, 아니면 음성클래스로 구분한다.  

## 오즈비
시그모이드 함수는 오즈비라는 통계를 기반으로 만들어지며  
다음과 같이 정의할 수 있다.  


```
OR = p / (p-1)

p = 성공확률
```

## 로짓함수
오즈비에 로그 함수를 취해서 만든 함수다.  
다음과 같이 정의할 수 있다.

```
logit(p) = log(OR)

OR은 선술한 오즈비
```

로짓 함수는 p가 0.5일 때 0이되고 0과 1일때 각각 무한대로 음수와 양수가 된다.  

이 식은 다음과 같이 쓸 수 있다.

```
log(p / (1 - p)) = z
```

## 로지스틱 함수

위 식을 다시 z에 대하여 정리하면 다음과 같은 식이 된다.

```
p = 1 / (1 + e ^(-1))
```
이게 로지스틱 함수다.  

로지스틱 함수를 그래프로 그리면 로짓 함수의 가로세로가 뒤집힌 형태다.  
로지스틱 함수를 시그모이드 함수라고도 부른다.

# 로지스틱 회귀 중간 정리

로지스틱 회귀는 이진분류가 목표다.  
때문에 무한대의 범위를 가지는 z값을 조절하기 위해  
시그모이드 함수를 활성화 함수로 사용한다.  
시그모이드 함수를 통과하면 확률처럼 해석할 수 있기 때문이다.  
  
그리고 그 결과를 임계함수에 넣어 0인지 1인지 구분한다.  
즉 이진 분류가 된 것이다.  
  
이제 가중치와 절편을 적절히 업데이트 하는 방법,  
로지스틱 회귀에 적용할 손실함수에 대해서 알아본다.

---

# 로지스틱 손실 함수
로지스틱 손실 함수는 크로스 엔트로피 손실 함수를 이진분류 버전으로 만든 것이다.  

이것을 식으로 나타내면 아래와 같다.  
a는 활성화 함수가 출력한 값이고 y는 타깃이다.  
```
L = -(ylog(a) + (1 - y)log(1 - a))
```

이진분류는 0과 1로 분류하는 함수로 답은 오직 둘 중 하나만 나온다.  
따라서 위 식은 y가 0이냐 1이냐로 정리된다.  
  
  
## 로지스틱 손실 함수 미분
  
로지스틱 손실함수를 미분한 식은 가중치와 절편 업데이트에 사용된다.

```
가중치의 경우 = - (y - a)x
절편의 경우   = - (y - a)1
```

로지스틱 손실함수의 미분을 통해 로지스틱 손실함수의 값을 최소로 하는  
가중치와 절편을 찾아야 한다.
